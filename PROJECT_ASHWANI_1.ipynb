```python
import pandas as pd
import matplotlib.pyplot as plt

matches = pd.read_csv(r"matches.csv")
deliveries = pd.read_csv(r"deliveries.csv")
```

```python
matches
```

```python
deliveries
```

```python
# i) Match won by the maximum margin of runs
matches.loc[matches['win_by_runs'].idxmax(), 
            ['season','team1','team2','winner','win_by_runs']]
```

```python
# ii) Match won by maximum wickets
matches.loc[matches['win_by_wickets'].idxmax(), 
            ['season','team1','team2','winner','win_by_wickets']]
```

```python
# iii) Match won by the minimum margin of runs
matches[matches['win_by_runs'] > 0].loc[
    matches[matches['win_by_runs'] > 0]['win_by_runs'].idxmin(),
    ['season','team1','team2','winner','win_by_runs']
]
```

```python
# iv) Match won by minimum wickets
matches[matches['win_by_wickets'] > 0].loc[
    matches[matches['win_by_wickets'] > 0]['win_by_wickets'].idxmin(),
    ['season','team1','team2','winner','win_by_wickets']
]
```

```python
# v) Matches where D/L method was and wasn't applied
matches['dl_applied'].value_counts()
```

```python
# vi) No. of matches held in each city
matches['city'].value_counts()
```

```python
# vii) No. of matches won by each team
matches['winner'].value_counts()
```

```python
matches['winner'].value_counts().plot(kind='bar', colormap='viridis')
```

```python
# viii) No. of matches held every season
matches['season'].value_counts().sort_index()
```

```python
# ix) Top 10 players based on no. of Man of Match (MOM) awards won
matches['player_of_match'].value_counts().head(10)
```

```python
# x) Does winning the toss mean winning the match?
toss_win = matches[matches['toss_winner'] == matches['winner']].shape[0]
total_matches = matches.shape[0]
(toss_win / total_matches) * 100
```

```python
# xi) What was the decision taken by captains when they won the toss?
matches['toss_decision'].value_counts()
```

```python
# xii) No. of matches where D/L method was applied every season
matches[matches['dl_applied'] == 1]['season'].value_counts().sort_index()
```

```python
# xiii) Different results for matches
matches['result'].value_counts()
```

```python
# xiv) How many times did a team win the toss?
matches['toss_winner'].value_counts()
```

```python
matches['toss_winner'].value_counts().plot(kind='barh', colormap='viridis')
```

```python
# xv) Best venue for defending a total
matches[matches['win_by_runs'] > 0]['venue'].value_counts().head(1)

# Best venue for chasing a total
matches[matches['win_by_wickets'] > 0]['venue'].value_counts().head(1)
```

```python
# xvi) Best defending team
matches[matches['win_by_runs'] > 0]['winner'].value_counts().head(1)

# Best chasing team
matches[matches['win_by_wickets'] > 0]['winner'].value_counts().head(1)
```

```python
# xvii) No. of matches played in different stadiums
matches['venue'].value_counts()
```
